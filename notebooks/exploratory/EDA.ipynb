```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_train = pd.read_csv('../../data/processed/train.csv')
df_test = pd.read_csv('../../data/processed/test.csv')
```

# Proporção de labels

```python
# Supondo que df_train e df_test possuem rótulos de 0 a 9 na coluna 'label'
# Vamos mapear esses rótulos para seus respectivos nomes usando a lista label_names.

label_names = ["airplane", "automobile", "bird", "cat", "deer", 
               "dog", "frog", "horse", "ship", "truck"]

# Mapear os rótulos numéricos para os nomes correspondentes nos dataframes
df_train['label_name'] = df_train['label'].map(lambda x: label_names[x])
df_test['label_name'] = df_test['label'].map(lambda x: label_names[x])

# Contagem dos rótulos para cada dataframe usando os nomes
train_label_counts = df_train['label_name'].value_counts()
test_label_counts = df_test['label_name'].value_counts()

# Combinando os dados de contagem para visualização
combined_counts = pd.DataFrame({
    'Train': train_label_counts,
    'Test': test_label_counts
}).fillna(0)

# Plotando o gráfico de barras com as quantidades absolutas mapeadas
combined_counts.plot(kind='bar', figsize=(12, 7))

plt.title('Quantidade Absoluta dos Rótulos no df_train e df_test')
plt.ylabel('Quantidade')
plt.xlabel('Rótulo')
plt.xticks(rotation=45)
plt.legend(title='Dataframe')

plt.show()
```

```python
# Supondo que df_train e df_test já estão definidos e possuem rótulos de 0 a 9 na coluna 'label'.
# Vamos associar os rótulos numéricos aos seus respectivos nomes.

label_names = ["airplane", "automobile", "bird", "cat", "deer", 
               "dog", "frog", "horse", "ship", "truck"]

# Mapear os rótulos numéricos para os nomes correspondentes nos dataframes
df_train['label_name'] = df_train['label'].map(lambda x: label_names[x])
df_test['label_name'] = df_test['label'].map(lambda x: label_names[x])

# Contagem dos rótulos para cada dataframe usando os nomes
train_counts = df_train['label_name'].value_counts(normalize=True)
test_counts = df_test['label_name'].value_counts(normalize=True)

# Combinar os dados para visualização
combined_counts = pd.DataFrame({
    'Train': train_counts,
    'Test': test_counts
}).fillna(0)

# Plotar o gráfico de colunas
combined_counts.plot(kind='bar', figsize=(12, 7))

plt.title('Proporção entre Rótulos no df_train e df_test')
plt.ylabel('Proporção')
plt.xlabel('Rótulo')
plt.xticks(rotation=45)
plt.legend(title='Dataframe')

plt.show()
```
