```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df_train = pd.read_csv('../../data/processed/train.csv')
#df_val = pd.read_csv('../../data/processed/val.csv')
df_test = pd.read_csv('../../data/processed/test.csv')
```

```python
def convert_to_image(df):
    # Separe os canais de cores
    red = df.iloc[:, :1024].values.reshape(-1, 32, 32)
    green = df.iloc[:, 1024:2048].values.reshape(-1, 32, 32)
    blue = df.iloc[:, 2048:3072].values.reshape(-1, 32, 32)
    
    # Combine os canais para criar a imagem
    images = np.stack([red, green, blue], axis=3)
    return images

# Convertendo o DataFrame para imagens
train_images = convert_to_image(df_train)
#val_images = convert_to_image(df_val)
test_images = convert_to_image(df_test)
```

```python
def plot_images(images, labels, label_names, num_images=10):
    plt.figure(figsize=(10, 10))
    for i in range(num_images):
        plt.subplot(1, num_images, i + 1)
        plt.imshow(images[i])
        plt.title(label_names[labels[i]])
        plt.axis('off')
    plt.show()
```

```python
# Exemplo de nomes das classes (substitua por sua lista de labels se necessrio)
label_names = ["airplane", "automobile", "bird", "cat", "deer", 
               "dog", "frog", "horse", "ship", "truck"]

# Visualizando as primeiras 10 imagens do conjunto de treinamento
plot_images(train_images, df_train['label'].values, label_names, num_images=10)

#plot_images(val_images, df_val['label'].values, label_names, num_images=10)

# Visualizando as primeiras 10 imagens do conjunto de teste
plot_images(test_images, df_test['label'].values, label_names, num_images=10)
```

# Visualizao por classe

```python
# Lista de nomes das classes
label_names = ["airplane", "automobile", "bird", "cat", "deer", 
               "dog", "frog", "horse", "ship", "truck"]

# Criando DataFrames por classe para o conjunto de treino
dfs_train = {label: df_train[df_train['label'] == i] for i, label in enumerate(label_names)}

# Criando DataFrames por classe para o conjunto de teste
dfs_test = {label: df_test[df_test['label'] == i] for i, label in enumerate(label_names)}
```

```python
# Visualizando imagens de treino
for label in label_names:
    print(f"Visualizando imagens da classe '{label}' no conjunto de treino")
    plot_images(convert_to_image(dfs_train[label]), dfs_train[label]['label'].values, label_names, num_images=10)

# Visualizando imagens de teste
for label in label_names:
    print(f"Visualizando imagens da classe '{label}' no conjunto de teste")
    plot_images(convert_to_image(dfs_test[label]), dfs_test[label]['label'].values, label_names, num_images=10)
```
